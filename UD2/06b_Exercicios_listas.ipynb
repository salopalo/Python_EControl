## Exercicio 1

Crea unha lista chamada **abecedario** que conteña como elementos as letras do abecedario español (incluíndo o "ñ") - 27 letras - 

Utiliza a constante do módulo string que contén o abecedario inglés para crear a lista por comprension e despois o método insert() para engadir o "ñ" á lista creada.

```python
import string

abecedario =[x for x in string.ascii_lowercase]
abecedario.insert(14, "ñ")
print(abecedario)
print(len(abecedario))
```

## Exercicio 2

Crea unha variable chamada **abc** de tipo string que conteña as letras do abecedario español (incluíndo o "ñ")  - 27 letras - 

Utiliza o operador + de concatenación e o slicing sobre a constante correspondente do módulo string.

A partir da variable **abc**, crea unha lista por comprensión chamada **vogais_ascii** na que se inclúan os números ASCII das letras que son vogais.


```python
import string

abc = string.ascii_lowercase[:14] + "ñ" + string.ascii_lowercase[14:]
print(abc)

vogais_ascii = [ ord(x) for x in abc if x in("aeiou")]
print(vogais_ascii)
```

## Exercicio 3

Fai o mesmo de antes pero construíndo o resultado como string en lugar de como lista. Chámalle **texto2**.


```python
texto = "Python"
cifrado = [chr(ord(x)+5) for x in texto]
print (cifrado)

texto2 = "".join(cifrado)
texto3 = ""
for x in cifrado:
    texto3 += x

texto4 = ""
for x in texto:
    texto4 += chr(ord(x) + 5)
    
texto5 = ""
for x in texto:
    texto5.join(chr(ord(x) + 5))
               

    
print(texto2)
print(texto3)
print(texto4)
print(texto5)
```
