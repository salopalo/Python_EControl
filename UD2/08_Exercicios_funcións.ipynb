## Exercicio 1

Realiza un programa que pida desde teclado os seguintes datos relativos a un contribuínte:

- Soldo bruto anual
- Tipo de IRPF ao que está suxeito
- Idade

A partir de estes datos calcularase o importe que lle corresponde como pago do IRPF a Facenda, en base ao seguinte:

O cálculo do imposto farase aplicando a porcentaxe do tipo IRPF sobre o soldo bruto, e sumando despois 100 euros máis a todos os contribuíntes, a excepcion dos maiores de 50 anos que no seu lugar recibirán unha bonificación de 170 euros na súa declaración.

Para finalizar mostrarase por pantalla unha mensaxe co importe total a pagar.

```python
soldo = float(input("Introduza o soldo: \n\n"))
irpf = float(input("Introduza o tipo de IRPF: \n\n"))
idade = int(input("Introduza a idade: \n\n"))

imposto = soldo * irpf/100
if idade <= 50:
    imposto += 100
else:
    imposto -= 170
    
print(f"O imposto a pagar é de : {imposto}€")



```

## Exercicio 3

Crea unha función chamada **calcula_IRPF** que reciba como argumentos:

- O soldo bruto anual do traballador (exemplo: 35722.50)
- O tipo de IRPF que se lle aplica (exemplo: 18.5)
- A idade do traballador (exemplo: 34)

E que devolva o importe total que lle corresponde pagar a Facenda.

O cálculo do imposto farase aplicando a porcentaxe do tipo IRPF sobre o soldo bruto, e sumando despois 100 euros máis a todos os traballadores, con excepcion dos maiores de 50 anos que recibirán unha bonificación de 170 euros na súa declaración.

Fai unha chamada a dita función cos datos indicados nos exemplos e comproba que o resultado que devolve nese caso é 6708.66 euros.


```python
def calcula_irpf(soldo, irpf, idade):
    imposto = soldo * irpf/100
    if idade <= 50:
        imposto += 100
    else:
        imposto -= 170
        
    return (f"O imposto a pagar é de : {imposto:.2f}€")

calcula_irpf(35722.50,18.5,34)
```
