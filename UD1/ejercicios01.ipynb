Crea un programa para resolver una ecuacion de segundo grado:

```python
print("Dame a:")
a = float(input())
print("Dame b:")
b = float(input())
print("Dame c:")
c = float(input())

x1 = (-b + (b**2 - 4*a*c)**0.5 ) / (2*a)
x2 = (-b - (b**2 - 4*a*c)**0.5 ) / (2*a)

print("Soluciones a la ecuacion: ", x1 , x2)
```
