## Celas de Notebook 

- Un Notebook é unha sucesión de "celas".
- A cela activa ten un borde darredor dela.
- Pódeste situar nas distintas celas mediante as teclas do cursor.
- Unha cela podes ser de código, de markdown ou "en bruto".
    - Código
        - Python
        - R
        - Julia
        - [Moitos outros](https://github.com/jupyter/jupyter/wiki/Jupyter-kernels)
- Pódense crear celas nun Notebook mediante o menú ou usando un atallo de teclado:
    - `a`: Insertar cela

## Modo de edición

|             |Modo de edición|Modo de execución|
|--|--|--|
|Como se entra | `Enter`,<br> ou doble-click | `Esc` ou `Ctrl-m` | 
|Borde da cela  |Azul interno    |Azul externo        |
|Acción       | Escribir código ou texto | Accións de nivel de Notebook |

```python
a = 3
```

## Tipos de cela

Cando se crea unha cela é sempre de tipo "código". Para convertila a Markdown ou "en bruto" hai que cambiar o tipo no menú da barra de ferramentas. Observa a diferencia cambiando o tipo da seguinte cela.

```python
# Heading 1
## Heading 2

2*10*1.5
```

## Algúns atallos de teclado

* `Ctrl`-`Enter`: Executar cela
* `Shift`-`Enter`: Executar cela e cambiar á cela inferior.
* `Alt`-`Enter`: Executar cela e insertar unha nova cela debaixo.
* `a` : Insertar cela (arriba)
* `b` : Insertar cela (abaixo)
* `dd` : Borrar cela (igual que en VIM)
* `Ctrl`-`s`: Gardar Notebook
* `i`: Interrumpir kernel (presionar dúas veces).
* `0`: Rebotar kernel (presionar dúas veces).

## Execución ordeada

Cada cela de código mostra á esquerda a orde secuencial na que foi executada. É dicir, a primeira cela en ser executada mostra un [1]

A orde na que as celas se executan importa. Proba a executar estas dúas celas...

```python
a = 1
b = 3
```

```python
a = a * b
print('a =',a)
print('b =',b)
```

Se executas varias veces esta última cela os resultados cambian. 

Isto pódese evitar correndo o notebook enteiro, mediante a opción de menú:

Kernel > "Restart Kernel and Run All Cells"
