# Introdución a Python

##  Breve historia e algúns feitos

* Creada por Guido Van Rossum, quen ostentou o cargo de "Dictador Benevolente Vitalicio" (BDFL) ata o ano 2018, no que se creou o Consello Rector de Python (Python Steering Council)
* A súa implementación comenzou en decembro de 1989, a primeira publicación apareceu en febreiro de 1991
* Tén influencias de Modula-3, Lisp, C, e outras linguaxes de programación

* Python 2.0 publicouse no ano 2000
* Novedades máis importantes: recollida de lixo e soporte para Unicode
* Tendencia a maior transparencia e desenvolvemento liderado pola comunidade
* Python 2 quedou obsoleto e discontinuado a partir da versión 2.7.18 in 2020

* Python 3.0 liberouse no 2008
* Supón unha adaptación ó modo "moderno de traballar".
* Como corrixiu erros importantes, non é compatible con versións anteriores (unha chea de código Python 2 non corre en Python 3)
* A versión máis moderna é Python 3.11 - maio 2023, pero calquera versión maior de 3.5 debería funcionar neste curso. Se hai problemas solucionarémolos a medida que se manifesten.

if n < 0:
    raise ValueError('You must enter a non negative integer')

for i in range(2, n + 1):
    factorial = factorial*i

print(factorial)
